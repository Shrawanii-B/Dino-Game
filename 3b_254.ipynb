# Theory
The Iris dataset is a classic dataset in machine learning containing 150 samples of iris flowers from three species: **setosa, versicolor, and virginica**. Each sample has four features: sepal_length, sepal_width, petal_length, petal_width.

Data visualization helps:

- Detect **patterns and relationships** between features.
- Compare feature distributions across **different species**.
- Provide insights for **classification tasks**.

##Code and Output:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

sns.set(style="whitegrid")
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns = df.columns.str.strip().str.lower()
print(df.columns.tolist())
```

```python
# Sepal Length Distribution
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='sepal length (cm)', hue='species', bins=20, kde=True, alpha=0.6, palette='Set1')
plt.title("Sepal Length Distribution by Species")
plt.show()

# Sepal Width Distribution
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='sepal width (cm)', hue='species', bins=20, kde=True, alpha=0.6, palette='Set2')
plt.title("Sepal Width Distribution by Species")
plt.show()

# Petal Length Distribution
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='petal length (cm)', hue='species', bins=20, kde=True, alpha=0.6, palette='Set3')
plt.title("Petal Length Distribution by Species")
plt.show()

# Petal Width Distribution
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='petal width (cm)', hue='species', bins=20, kde=True, alpha=0.6, palette='pastel')
plt.title("Petal Width Distribution by Species")
plt.show()
```

```python
features = df.columns[:-1]  # all numerical features except 'species'

for col in features:
    plt.figure(figsize=(7,4))
    ax = sns.boxplot(x='species', y=col, data=df, hue='species', palette='Set2')
    plt.title(f"{col.title()} by Species")
    plt.show()
```

```python
sns.pairplot(df, hue='species', palette='bright')
plt.suptitle("Pairwise Feature Relationships by Species", y=1.02)
plt.show()
```

##Observations:
1. Sepal Length Histogram: Setosa has smaller sepal lengths; Versicolor and Virginica overlap moderately.

2. Sepal Width Histogram: Setosa generally has wider sepals; Virginica has narrower sepals.

3. Petal Length Histogram: Setosa petals are very small; Versicolor and Virginica overlap slightly.

4. Petal Width Histogram: Setosa has narrow petals; Versicolor and Virginica overlap with Virginica slightly wider.

5. Boxplots: Petal length and width clearly separate Setosa; Versicolor and Virginica show partial overlap in sepal features.

6. Pairplot: Petal length and width are the most discriminative features; Setosa is clearly separated from other species.
